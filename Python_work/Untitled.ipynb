```python
def get_MC_simulation(close_prices_df, client_portfolio):
    if investment_length == "1yr":
        MC_oneyear =  MCSimulation(
        portfolio_data=close_prices_df,
        weights=client_portfolio,
        num_simulation=500,
        num_trading_days=252,
    )
        MC_cum = MC_oneyear.calc_cumulative_return()
        MC_cum_mean = MC_cum.mean(axis=1)
        MC_sim_line_plot = MC_oneyear.plot_simulation()
        MC_sim_line_plot.get_figure().savefig("img/mc.png", bbox_inches="tight")
    
    elif investment_length == "3yr":
        MC_twoyear =  MCSimulation(
        portfolio_data=close_prices_df,
        weights=client_portfolio,
        num_simulation=500,
        num_trading_days=252*2,
    )
        MC_cum = MC_twoyear.calc_cumulative_return()
        MC_cum_mean = MC_cum.mean(axis=1)
        MC_sim_line_plot = MC_twoyear.plot_simulation()
        MC_sim_line_plot.get_figure().savefig("img/mc.png", bbox_inches="tight")
    
    elif investment_length == "1yr":
        MC_threeyear =  MCSimulation(
        portfolio_data=close_prices_df,
        weights=client_portfolio,
        num_simulation=500,
        num_trading_days=252*3,
    )
        MC_cum = MC_threeyear.calc_cumulative_return()
        MC_cum_mean = MC_cum.mean(axis=1)
        MC_sim_line_plot = MC_threeyear.plot_simulation()
        MC_sim_line_plot.get_figure().savefig("img/mc.png", bbox_inches="tight")
    
    return MC_cum, MC_cum_mean
```
